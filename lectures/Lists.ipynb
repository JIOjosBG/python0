Разпакетиране

```python
first, second, *other = letters
first, second, other
```

Обхождане

```python
for letter in letters:
    print(letter)
```

```python
for index, letter in enumerate(letters):
    print(index, letter)
```

Добавяне на елемент

```python
letters.append("e")
letters
```

```python
letters.insert(0, "-")
letters
```

Премахване на елемент

```python
last = letters.pop()
last, letters
```

```python
first = letters.pop(0)
first, letters
```

```python
letters.remove("b")
letters
```

```python
del letters[0:3]
letters
```

Търсене елемент по стойност

```python
letters = ["a", "b", "c"]
if "f" in letters:
    print('f is in letters')
```

```python
letters.index("f")
```

```python
letters.index("a")
```

Подреждане

```python
letters.sort()
letters
```

```python
letters.sort(reverse=True)
letters
```
